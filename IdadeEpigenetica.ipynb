## Encontrar, no nosso banco de dados com os CpGs, todos os sítios necessários para o relógio epigenético (*match*)


```r
match1=match(datMiniAnnotation[,1], met$ProbeID)

#sítios CpG não encontrados aparecem como "NA"
```

## Manter, no nosso arquivo de betas, apenas os CpGs necessários para o relógio epigenético (lista match1)

```r
head(met)
```

```r
met_reduced=met[match1,]

#ou: met_reduced=met[match1,]
```

```r
dim(met_reduced)
```

```r
met_reduced[,7]=as.character(met_reduced[,7])
#met_reduced$ProbeID=as.character(met_reduced$ProbeID)

#converte os valores da 7ª coluna de met_reduced (nomes dos CpGs) para o tipo "character"
```

```r
met_reduced[is.na(match1),7]=as.character(datMiniAnnotation[is.na(match1),1])

#substitui os valores da primeira coluna de met_reduced pelos valores correspondentes da primeira coluna de datMiniAnnotation, mas apenas nas linhas onde match1 é NA.

met_out=data.frame(met_reduced)
```

```r
head(met_out)
```

```r
#make sure you output numeric variables...
for (i in 2:(dim(met_out)-1)[[2]] ){met_out[,i]=as.numeric(as.character(gsub(x=met_out[,i],pattern="\"",replacement=""))) }

#itera sobre todas as linhas de met_out, exceto a primeira, removendo aspas dos valores e convertendo-os para números

met_out = met_out[,c(7,2:6)]
#colnames(met_out)[1] = "ProbeID"

write.table(met_out,"MethylationData.csv", row.names=F, sep=",")
```
